importing nescessary modules

```python
import pandas as pd
import pandas_ta as ta
```

taking the raw data set from the given link

```python
url = "https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=IBM&interval=5min&apikey=demo"

df = pd.read_json(url)
df.head(30)
```

splited the given data set into metadata.json and time.json,
metadata consist of company details,
time.json consists of the time intervals of the market

reading the metadata.json file

```python
df1 = pd.read_json("metadata.json")
df1.head(30)
```

reading the time.json file

```python
df2 = pd.read_json("time.json")
df2 = df2.transpose()
df2.head(30)
df2 = df2.rename(columns={"1. open":"open" , "2. high":"high", "3. low":"low" , "4. close":"close" , "5. volume":"volume"})
df2.head()
```

adding ATR_PERIOD and MULTIPLIER according to the given question

```python
ATR_PERIOD = 7
MULTIPLIER = 3
```

supertrending the data using the python module named pandas_ta
which uses the folowing formula:-

BASIC UPPERBAND = (HIGH + LOW) / 2 + Multiplier * ATR
BASIC LOWERBAND = (HIGH + LOW) / 2 - Multiplier * ATR

```python
ta.supertrend(high=df2['high'],low=df2['low'],close=df2['close'],period = ATR_PERIOD, multiplier=MULTIPLIER)
```

adding an extra column in our dataframe to store the values of SUPERT_7_3.0 for further process and printing the final dataframe

```python
df2['sup'] = ta.supertrend(high=df2['high'],low=df2['low'],close=df2['close'],period = ATR_PERIOD, multiplier=MULTIPLIER)['SUPERT_7_3.0']
```

```python
df2.head(30)
```

adding two more column in our data frame to indicate when to buy and when to sell and printing the final dataframe

```python
df2['Buy_stuff'] = 0
df2['Sell_stuff'] = 0
```

```python
df2
```

running a simple for loop with initial condition ATR_PERIOD-1 to length of our dataframe which will decide whether to buy stuffs or sell stuffs

```python

for i in range(ATR_PERIOD-1,len(df2)):
    if df2['close'][i-1]<=df2['sup'][i-1] and df2['close'][i] > df2['sup'][i]:
        df2['Buy_stuff'][i] = 1
    if df2['close'][i-1]>=df2['sup'][i-1] and df2['close'][i] < df2['sup'][i]:
        df2['Sell_stuff'][i] = 1
```

printing only the time frames where you can either buy or sell stuffs ('1' indicate to 'Yes' and '0' indicate to 'No')

```python
df2[(df2['Buy_stuff'] > 0) | (df2['Sell_stuff'] > 0)]
```
